Q3

```python
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.to_period('M')
monthly_prices = df.groupby(['month', 'category'])['usdprice'].mean().reset_index()
price_pivot = monthly_prices.pivot(index='month', columns='category', values='usdprice')
volatility = price_pivot.std()
volatility_sorted = volatility.sort_values(ascending=False)
volatility_sorted
```

cereals and tubers have the higest price volatility
